```python
x = [0,2,3,4,5]
y = [1, 4, 9, 16, 25]
fig, ax = plt.subplots(2, 2)
ax = ax.ravel()
ax[0].plot(x, y)
ax[1].plot(y,x)
ax[2].plot(x, x)
ax[3].plot(y,y)
```

```python
x = [0,2,3,4,5]
y = [1, 4, 9, 16, 25]
fig, ax = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.4, wspace=0.4)
ax = ax.ravel()
ax[0].plot(x, y)
ax[1].plot(y,x)
ax[2].plot(x, x)
ax[3].plot(y,y)

xlabel = ["X", "Y", "X", "Y"]
ylabel = ["Y", "X", "X", "Y"]
for i, ax0 in enumerate(ax):
  ax0.set_xlabel(xlabel[i])
  ax0.set_ylabel(xlabel[i])

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib

df = sns.load_dataset('tips')
df.head()
```

```python
fig, ax = plt.subplots()

ax = sns.scatterplot(x="total_bill", y="tip", data=df)
ax.set_title("支払額とチップ")
ax.set_xlabel('支払額($)')
ax.set_ylabel('チップ($)')
```

```python
fig, ax = plt.subplots()

ax = sns.scatterplot(x="total_bill", y="tip", data=df, hue="time")
ax.set_title("支払額とチップ")
ax.set_xlabel('支払額($)')
ax.set_ylabel('チップ($)')
```

```python
fig, ax = plt.subplots()

ax = sns.scatterplot(x="total_bill", y="tip", data=df, style="time")
ax.set_title("支払額とチップ")
ax.set_xlabel('支払額($)')
ax.set_ylabel('チップ($)')
```

```python
fig, ax = plt.subplots()

ax = sns.scatterplot(x="total_bill", y="tip", data=df, hue="time", alpha=0.5)
ax.set_title("支払額とチップ")
ax.set_xlabel('支払額($)')
ax.set_ylabel('チップ($)')
```

```python
fig, ax = plt.subplots()

ax = sns.scatterplot(x="total_bill", y="tip", data=df, hue="time", alpha=0.5, size="size",
                     sizes=(10, 500))
ax.set_title("支払額とチップ")
ax.set_xlabel('支払額($)')
ax.set_ylabel('チップ($)')
```
